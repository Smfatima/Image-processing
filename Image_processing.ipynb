Thresholding using OpenCV

```
import cv2 as cv
img = cv.imread('abu.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#Simple thrsholding
# threshold, thresh = cv.threshold(gray, 100, 255, cv.THRESH_BINARY)

#Threshold inverse
# threshold, thresh = cv.threshold(gray, 100, 255, cv.THRESH_BINARY_INV)

#Adpotive thresholding

ado_thrs =  cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,9)

plt.imshow(ado_thrs)
plt.show()
```

### Filters!

Edge detection using Sobel

```
from skimage.filters import sobel
img = cv.imread('abu.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#Apply edge detection
edge_sobe = sobel(gray)
plt.imshow(edge_sobe)
plt.show()
```

### Edge detection using Canny

```
from skimage.feature import canny

#grayscale picture
img = cv.imread('abu.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#canny edges
cany_ed = canny(gray, sigma=1.3)
plt.imshow(cany_ed)
plt.show()
```

Gaussian Smoothing

```
#import the module
from skimage.filters import gaussian

img = plt.imread('abu.jpg')
#apply edge detection filter
guass = gaussian(img, multichannel=True)

plt.imshow(guass)
plt.show()
```

### Removing noise 

```
#Import 
from skimage.restoration import denoise_bilateral

#apply it
de_ny = denoise_bilateral(gray, multichannel=False)

plt.imshow(de_ny)
plt.show()
```

## Super pixel segmentation

```
from skimage.segmentation import slic
from skimage.color import label2rgb

img = plt.imread('abu.jpg')
segments = slic(img, n_segments=900)

seg_img = label2rgb(segments, img, kind='avg')
plt.imshow(seg_img)
plt.show()
```
